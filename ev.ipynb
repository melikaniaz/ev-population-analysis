```python
import matplotlib.pyplot as plt

# group by Model Year
trend = df.groupby("Model Year").size().reset_index(name="count")

# show first rows
print(trend.head())

# plot trend
plt.figure(figsize=(10,5))
plt.plot(trend["Model Year"], trend["count"], marker="o")
plt.title("EV Registrations Over Time")
plt.xlabel("Year")
plt.ylabel("Number of Vehicles")
plt.grid(True)
plt.show()
```

```python
# Top 10 brands by number of EVs
top_brands = df["Make"].value_counts().head(10)
print(top_brands)

# Plot bar chart
plt.figure(figsize=(10,5))
top_brands.plot(kind="bar")
plt.title("Top 10 EV Brands")
plt.xlabel("Brand")
plt.ylabel("Number of Vehicles")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Count EV types
type_counts = df["Electric Vehicle Type"].value_counts()
print(type_counts)

# Plot pie chart
plt.figure(figsize=(6,6))
type_counts.plot(kind="pie", autopct="%1.1f%%", startangle=90)
plt.title("EV Types: BEV vs PHEV")
plt.ylabel("")  # remove y-label
plt.show()
```

```python
# Drop rows with missing or zero values for clean analysis
df_range_price = df.dropna(subset=["Electric Range", "Base MSRP"])
df_range_price = df_range_price[(df_range_price["Electric Range"] > 0) & (df_range_price["Base MSRP"] > 0)]

print("Shape after cleaning for range/price:", df_range_price.shape)

# 1) Distribution of Electric Range
plt.figure(figsize=(8,5))
df_range_price["Electric Range"].plot(kind="hist", bins=40)
plt.title("Distribution of Electric Range")
plt.xlabel("Range (miles)")
plt.ylabel("Frequency")
plt.show()

# 2) Distribution of Base MSRP
plt.figure(figsize=(8,5))
df_range_price["Base MSRP"].plot(kind="hist", bins=40)
plt.title("Distribution of Base MSRP")
plt.xlabel("Price (USD)")
plt.ylabel("Frequency")
plt.show()

# 3) Scatter plot Range vs Price
plt.figure(figsize=(8,6))
plt.scatter(df_range_price["Electric Range"], df_range_price["Base MSRP"], alpha=0.5)
plt.title("Electric Range vs Base MSRP")
plt.xlabel("Range (miles)")
plt.ylabel("Price (USD)")
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from pptx import Presentation
from pptx.util import Inches

# Load cleaned data
df = pd.read_csv("EV_clean.csv")

# Prepare PowerPoint
prs = Presentation()
title_slide_layout = prs.slide_layouts[0]
slide = prs.slides.add_slide(title_slide_layout)
slide.shapes.title.text = "Electric Vehicle Population Analysis"
slide.placeholders[1].text = "Auto-generated via Python"

# --- 1. Trend by year ---
trend = df.groupby("Model Year").size().reset_index(name="count")
plt.figure()
plt.plot(trend["Model Year"], trend["count"], marker="o")
plt.title("EV Registrations Over Time")
plt.xlabel("Year")
plt.ylabel("Number of Vehicles")
plt.grid(True)
plt.savefig("trend.png", bbox_inches="tight")

slide = prs.slides.add_slide(prs.slide_layouts[5])
slide.shapes.title.text = "EV Registrations Over Time"
slide.shapes.add_picture("trend.png", Inches(1), Inches(1.5), width=Inches(8))

# --- 2. Top brands ---
top_brands = df["Make"].value_counts().head(10)
plt.figure()
top_brands.plot(kind="bar")
plt.title("Top 10 EV Brands")
plt.xlabel("Brand")
plt.ylabel("Number of Vehicles")
plt.savefig("brands.png", bbox_inches="tight")

slide = prs.slides.add_slide(prs.slide_layouts[5])
slide.shapes.title.text = "Top 10 EV Brands"
slide.shapes.add_picture("brands.png", Inches(1), Inches(1.5), width=Inches(8))

# --- 3. EV Types ---
type_counts = df["Electric Vehicle Type"].value_counts()
plt.figure()
type_counts.plot(kind="pie", autopct="%1.1f%%", startangle=90)
plt.title("EV Types: BEV vs PHEV")
plt.ylabel("")
plt.savefig("types.png", bbox_inches="tight")

slide = prs.slides.add_slide(prs.slide_layouts[5])
slide.shapes.title.text = "EV Types: BEV vs PHEV"
slide.shapes.add_picture("types.png", Inches(1), Inches(1.5), width=Inches(8))

# --- 4. Range vs Price ---
df_range_price = df.dropna(subset=["Electric Range", "Base MSRP"])
df_range_price = df_range_price[(df_range_price["Electric Range"] > 0) & (df_range_price["Base MSRP"] > 0)]
plt.figure()
plt.scatter(df_range_price["Electric Range"], df_range_price["Base MSRP"], alpha=0.5)
plt.title("Electric Range vs Base MSRP")
plt.xlabel("Range (miles)")
plt.ylabel("Price (USD)")
plt.grid(True)
plt.savefig("range_price.png", bbox_inches="tight")

slide = prs.slides.add_slide(prs.slide_layouts[5])
slide.shapes.title.text = "Electric Range vs Base MSRP"
slide.shapes.add_picture("range_price.png", Inches(1), Inches(1.5), width=Inches(8))

# Save PowerPoint
prs.save("EV_Report.pptx")
print("âœ… PowerPoint created: EV_Report.pptx")
```
